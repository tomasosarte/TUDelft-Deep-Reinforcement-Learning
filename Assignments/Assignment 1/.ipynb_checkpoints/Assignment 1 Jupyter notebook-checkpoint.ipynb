```python
import torch as th
import torchvision.datasets as datasets
import torchvision.transforms as transforms
from torch.utils.data import DataLoader

import numpy as np
import matplotlib.pyplot as plt
```

# Collecting Training & testing data

```python
data_train = datasets.MNIST(root='./data', train=True, download=True, transform = transforms.ToTensor())
train_set = DataLoader(data_train, batch_size=1024, shuffle=True)

data_set = datasets.MNIST(root='./data', train=False, download=True, transform = transforms.ToTensor())
test_set = DataLoader(data_set, batch_size=1024, shuffle=True)
```

# Function definition

```python
def train_and_test(model, train_set, test_set, criterion, optimizer, max_epochs, one_hot = False, 
                   flatten = False, float_y = False, transformation = False):
    train_loss, train_acc = [], []
    test_loss, test_acc = [], []

    for e in range(max_epochs):
        losses, accuracies = [], []
        test_losses, test_accuracies = [], []
        
        # TRAINING SET
        for x, y in train_set:
            mx = model(x)
            
            if transformation:
                mx = transformation_function(mx)
            if flatten:
                mx = th.flatten(mx)
            if float_y:
                y = y.float()
            if one_hot: 
                new_y = th.nn.functional.one_hot(y).float()
                loss = criterion(mx, new_y)
            else: 
                loss = criterion(mx, y)
            
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            losses.append(loss.item())
            acc = (mx.max(dim=-1)[1] == y).sum().item() / y.shape[0]
            accuracies.append(acc)

        train_loss.append(np.mean(losses))
        train_acc.append(np.mean(accuracies))
        
        # TESTING SET
        for x, y in test_set:
            mx = model(x)
            
            if transformation:
                mx = transformation_function(mx)
            if flatten:
                mx = th.flatten(mx)
            if float_y:
                y = y.float()
            if one_hot: 
                new_y = th.nn.functional.one_hot(y).float()
                loss = criterion(mx, new_y)
            else: 
                loss = criterion(mx, y)
                
            test_losses.append(loss.item())
            acc = acc = (mx.max(dim=-1)[1] == y).sum().item() / y.shape[0]
            test_accuracies.append(acc)
        
        test_loss.append(np.mean(test_losses))
        test_acc.append(np.mean(test_accuracies))
        
    return train_loss, train_acc, test_loss, test_acc
        
    
def plot_results(loss, acc, title):
    epochs = range(1, len(train_loss) + 1)

    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(epochs, loss, label= title + ' Loss', marker='o')
    plt.title(title + ' Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(epochs, acc, label= title + ' Accuracy', marker='o')
    plt.title(title + ' Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.suptitle(title + ' Results')
    plt.show()

def transformation_function(tensor, alpha = 4.5, beta = 4.5):
    return alpha * tensor + beta
```

# A1.1 a)

Here we define: the model, the criterion for the exercise and the optimizatzer used in the model.

```python
model = th.nn.Sequential(th.nn.Conv2d(1, 8, 5), # ( 8, 24, 24)
                         th.nn.ReLU(), # ( 8, 24, 24)
                         th.nn.MaxPool2d(2, 2), # ( 8, 12, 12)
                         th.nn.Conv2d(8, 16, 5), # (16, 8, 8)
                         th.nn.ReLU(), # (16, 8, 8)
                         th.nn.MaxPool2d(2, 2), # (16, 4, 4)
                         th.nn.Flatten(), # (16 * 4 * 4)
                         th.nn.Linear(256, 128), # (128)
                         th.nn.ReLU(), # (128)
                         th.nn.Linear(128, 32), # ( 32)
                         th.nn.ReLU(), # ( 32)
                         th.nn.Linear(32, 10)) # ( 10)

crossEntropyCriterion = th.nn.CrossEntropyLoss()
optimizer = th.optim.RMSprop(model.parameters(), lr=0.0005)
```

Get the results of train and test:

```python
train_loss, train_acc, test_loss, test_acc = train_and_test(model, train_set, test_set, crossEntropyCriterion, 
                                                            optimizer, max_epochs = 10)
```

Plot results:

```python
plot_results(train_loss, train_acc, 'Training')
plot_results(test_loss, test_acc, 'Testing')
```

Plot Results:

# A1.2 a) & b)

In order to prove questions in a) and b) it is usefull to develop de equation of the expectation of the formula:

- $ f_t = f_{t-1} + \alpha * (y_t - ) f_{t-1} $

- $ E[f_t] = E[f_{t-1}] + \alpha E[(y_t - ) f_{t-1}] $
- $ E[f_t] = E[f_{t-1}] + \alpha (\mu - E[f_{t-1}]) $
- $ E[f_t] = (1 - \mu)E[f_{t-1}] + \alpha\mu $ . <br> <br>
If we apply this rule recursively: <br> <br>
- $ E[f_t] = (1 - \mu)((1 - \mu)E[f_{t-2}] + \alpha\mu) + \alpha\mu $
- $ E[f_t] = (1 - \mu)^2E[f_{t-2}] + \alpha\mu + (1 - \mu)\alpha\mu $
- $ E[f_t] = (1 - \mu)^3E[f_{t-3}] + \alpha\mu + (1 - \mu)\alpha\mu + (1 - \mu)^2\alpha\mu $ <br> <br>
This tends to: <br> <br>
- $ E[f_t] = (1 - \mu)^t E[f_1] + \alpha\mu \sum_{k = 0}^{t} (1 - \mu)^k$
- $ E[f_t] = (1 - \mu)^t f_1 + \alpha\mu \frac{1 -(1 - \mu)^{t+1}}{\mu}$
- $ E[f_t] = (1 - \mu)^t f_1 + \mu - \mu(1 - \mu)^{t+1}$

As we can see, when $t\to\infty$ the first term is going to be 0 and also the second term so $E[f_t] = \mu$.

In the other hand if t is a small number the first and third terms have impact on the formula so $E[f_t] \neq \mu$

## A1.3 a)

Since the los function is the mean squared error and the linear function is $f(x) = a^Tx$ we have that the loss function is exactly: <br><br>
- $L = \frac{1}{n}\sum_{i=1}^{n}(a^Tx - y_i)^2$ <br><br>
In order to find wich is the best solution for $a$ we need to derive the loss function with respect to $a^T$:<br><br>
- $\frac{dL}{da^T} = \frac{2}{n}\sum_{i=1}^{n}(a^Tx_i - y_i)x_i$ <br><br>
Once we have the derivative, we should equal it to 0 and simplify the expression: <br><br>
- $\frac{2}{n}\sum_{i=1}^{n}(a^Tx_i - y_i)x_i = 0$
- $2(\sum_{i=1}^{n}(a^Tx_ix_i) - \sum_{i=1}^{n}(x_iy_i))= 0$
- $\sum_{i=1}^{n}(a^Tx_ix_i) = \sum_{i=1}^{n}(x_iy_i)$
- $a^T\sum_{i=1}^{n}(x_ix_i) = \sum_{i=1}^{n}(x_iy_i)$
- $a^T = \frac{\sum_{i=1}^{n}(x_iy_i)}{\sum_{i=1}^{n}(x_ix_i)}$

## A1.3 b)

In order to compare the expected losses we need first to compute expected loss of the mean squared error of both equations: the one regarding $y_i$ and the other regarding $\hat{y_i}$. It's almost straight forward to se that they are:  <br><br>
- $\frac{1}{n}\sum{i=1}_{n}(E[f(x_i)] - E[y_i])$
- $\frac{1}{n}\sum{i=1}_{n}(E[f(x_i)] - E[\hat{y_i}])$ <br><br>

Since the only difference is $E[y_i]$ and $E[\hat{y_i}]$ we have to prove that they are equal: <br><br>
- $E[\hat{y_i}] = E[y_i + \epsilon_i]$ 
- $E[\hat{y_i}] = E[y_i] + E[\epsilon_i]$ <br><br>

Since $E[\epsilon_i] = 0$ we can conclude that $E[\hat{y_i}] = E[y_i]$
